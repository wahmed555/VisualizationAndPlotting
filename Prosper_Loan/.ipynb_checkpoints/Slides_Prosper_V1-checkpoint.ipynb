## BorrowerAPR or Interest on Loans:

- The Borrower's Annual Percentage Rate (APR) for the loan is different from the interest rate. APR not only includes interest costs, but also fees related to a loan.
- Borrower's Annual Percentage Rate (APR) tells you how much a loan will cost in one year.
- The APR graph seems to be normal distributed with peak at 0.2 APR.
- There is a strange spike in the 0.35-0.37 bin which indicates a strangely popular fee rate. this may be for primarily higher risk borrowers ( we will investigate this claim). 

```python

plt.figure(figsize=(11.7,8.27))
ticks =list(np.arange(0, 0.6, 0.1))
labels = ['{:.2f}'.format(v) for v in ticks]
bins = np.arange(0, df_0['BorrowerAPR'].max()+.02, .015)
plt.hist(df_0['BorrowerAPR'].dropna().values, bins = bins,rwidth=0.8)
# plt.hist(data = df_0, x = 'BorrowerAPR', bins = bins)
plt.xticks(ticks, labels)
plt.xlabel('Borrower APR',fontsize=14)
plt.ylabel('Count',fontsize=14)
plt.title("Borrower APR Count",fontsize=14)
```

 ## Relationship between borrowers 'Prosper rating' and their 'Credit score'
 
 - A credit score is one of the key indicator of determining a person's likelihood of paying credit back, so as we climb the 'Prosper rating' categories, the 'credit score' of the borrowers also tend to increase.
 
 - Strange thing to notice is HR -prosper rating has a slightly higher median than E despite being a riskier category - 

```python

order_CreditGrade=['AA','A','B','C','D','E','HR']
order_CreditGrade_reverse=reversed(order_CreditGrade)
plt.figure(figsize=( 11.69, 8.27))
sns.boxplot(data = df_0, y = 'CreditScoreRangeAvg',x = 'ProsperRating (Alpha)',order=order_CreditGrade_reverse)

plt.xlabel('ProsperRating Alpha',fontsize=13)
plt.ylabel('CreditScoreRangeAvg',fontsize=13)
plt.title('Borrowers Prosper rating Vs credit score Vs Rating Category',fontsize=13)
# plt.xticks(rotation = 45)
```

## 'BorrowerAPR' vs 'ProsperRatings'

- Credit Score is one of the factors of Prospers rating categories.  These categories, along with some other financial information, determine the Annual Percentage Rate (interest+ Fee overheads) that  will apply to the borrowers loan. 

- As we move from High risk to lower risk (Prosper Rating) i.e. - from a HR to an AA rating - the APR for the borrower reduces drastically. In fact, looking at the results, it goes from a median APR of 35% for High Risk all the way to a median value of 9% for AA.

- The variation in APRs also decreases as the (Prosper Rating Categories) get less riskier as displayed by the decreasing size of the boxes in the boxplots when going from HR to AA.

- The surprising thing is 'High Risk' has most consistent APR af all other Prosper rating categories.

```python
plt.figure(figsize=(11.69, 8.27))
order_CreditGrade_reverse=reversed(order_CreditGrade)

sns.boxplot(data = df_0, y = 'BorrowerAPR', x = 'ProsperRating (Alpha)',order=order_CreditGrade_reverse)

plt.xlabel('ProsperRating Alpha',fontsize=13)
plt.ylabel('BorrowerAPR',fontsize=13)
plt.title(' Prosper rating Vs Annual Percentage Rate (APR) Vs Rating Category',fontsize=13)
```
