## 3.  Histrogram using Bar plot
A bar chart can also be a good choice for plotting histograms of bucketed numeric
values

#### Method 1 for histogram

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
l1=[]
for i in grades:
    l1.append(i//10*10)
h1=Counter(l1)
h1
```

```python
plt.bar(h1.keys(),h1.values(),color="red")
```

```python
plt.bar(h1.keys(),h1.values(),8)
```

####  METHOD 2 for histrogram

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

# making a lambda function called decile to round in decade eg 83,87,85,82=80
decile = lambda grade: grade // 10 * 10

# applying decile  function on grade and then
histogram = Counter(decile(grade) for grade in grades)
```

```python
histogram
```

```python
plt.bar(histogram.keys(),histogram.values(),8)
```

##### To shipt x -axis markes 4 position to the left

```python
plt.bar([x - 4 for x in h1.keys()],h1.values(),8)
```

#### METHOD#3 

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

# making a lambda function called decile to round in decade eg 83,87,85,82=80
decile = lambda grade: grade // 10 * 10 

histogram = Counter(decile(grade) for grade in grades)
# shift each bar to the left by 4, # give each bar its correct height
plt.bar([x - 4 for x in histogram.keys()], histogram.values(),8)
```

```python
# Axis of Plot defines its size

# so that the “0” and “100” bars are fully shown
# signature[-x,+x,-y,+y]

plt.axis([-5, 105, 0, 5])
```

## 4. Line Charts

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
```
